```python
# Upload data from GitHub to notebook's local drive
url = "https://raw.githubusercontent.com/GokuMohandas/practicalAI/master/data/titanic.csv"
response = urllib.request.urlopen(url)
html = response.read()
with open('titanic.csv' , 'wb') as f:
    f.write(html)
```

## Some notes about 'wb'
* it means write in binary mode.

In UNIX like macOS system, the 'wb' does not do anything different. Remember to open pure text file using 'r' 'w', to open other files, use 'rb' 'wb'.

## 'with' statement in python

f = open('sometxt.txt','w')<br>
f.write('hha')<br>
f.close()<br>



can be wriiten as:



with open('sometxt.txt', 'w') as f:<br>
    f.write('hha')


shorter and no 'f.close' needed, which is added automatically.

```python
# extract people who have paid more than 200 to purchase the ticket

expensive = a[a['fare']>400]
expensive.describe()
```

```python
a = a.dropna()
```

```python
a = a.reset_index()
```

```python
a.head()
```

```python
a = a.drop(['name', 'cabin', 'ticket'],axis =1 )
```

```python
a.head()
```

```python
# Map feature values

a['sex'] = a['sex'].map({"female": 0,"male":1}).astype(int)
```

```python
a.head()
```

```python
# Lambda expressions to create new features

def get_family_size(sibsp, parch):
    family_size = sibsp + parch
    return family_size
a['family_size'] = a[['sibsp','parch']].apply( lambda x: get_family_size(x["sibsp"], x['parch']), axis = 1)
```

```python
a.head()
```

```python
a.head(20)
```

```python
def zhenhuaqian(sex, fare):
    zhenhuaqian = sex * fare
    return zhenhuaqian

a['zhenhuaqian'] = a[['sex','fare']].apply(lambda x: zhenhuaqian(x['sex'],x['fare']),axis = 1)
```

```python
a.head()
## gentlemen should buy the tickets for its mate.
```

```python
# Saving the date to CSV

a.to_csv("new_file.csv", index = False)
```
